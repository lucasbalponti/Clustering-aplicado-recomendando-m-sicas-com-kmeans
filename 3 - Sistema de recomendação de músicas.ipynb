```python
import pandas as pd
import numpy as np
```

```python
# Lendo a base de dados já com os clusters (não será upada no github devido ao seu tamanho de 250MB)

base_clusters = pd.read_csv(r'Dados/cluster_por_musica.csv')
base_clusters.head()
```

```python
# Definindo uma música para sugerir recomendações
artista_exemplo = 'Ed Sheeran'
musica_exemplo = 'Shape of You'
```

```python
# Verificando o cluster dessa música:
cluster_exemplo = base_clusters.loc[((base_clusters['artists'] == artista_exemplo) & (base_clusters['name'] == musica_exemplo))]['cluster_pca'].values[0]
cluster_exemplo
```

```python
# Verificando as métricas da música exemplo:
metricas_exemplo = base_clusters.loc[((base_clusters['artists'] == artista_exemplo) & (base_clusters['name'] == musica_exemplo))].loc[:, :'611'].values.reshape(-1, 1)
```

```python
# Adicionando uma coluna ao dataframe que representará a distância euclidiana da música até a Shape of You
base_clusters['Distância Shape of You'] = [np.linalg.norm(metricas_exemplo - base_clusters.loc[i, :'611'].values.reshape(-1, 1)) for i in base_clusters.index]
```

```python
base_clusters.head()
```

```python
# Verificando as músicas mais "próximas" à música selecionada como exemplo:

base_clusters[['artists', 'name', 'cluster_pca', 'Distância Shape of You']].sort_values(by = ['Distância Shape of You'], ascending=True).head(10)
```
